```r
n = 100
means = NULL
upper_var = NULL
lower_var = NULL

h = function(x) {
    return ((cos(50*x) + sin(20*x))^2)
}

var = function(x,m) {
  var = (1/m)*mean((h(x) - mean(h(x)))^2)
  return (sqrt(var))
}


ms <- seq(100, 20000, by = 100)

for (m in ms) {

  x = runif(m)
  means = c(means, mean(h(x)))
  upper_var = c(upper_var, mean(h(x))+var(x,m))
  lower_var = c(lower_var, mean(h(x))-var(x,m))

}
```

```r
plot(ms, means, col = 'black')
lines(ms, lower_var, col = 'blue')
lines(ms, means, col = 'black')
lines(ms, upper_var, col = 'red')
```

```r
# question 3

h = function(x) {
  (cos(50*x) + sin(20*x))^2
}

v = function(x, m) {
  sqrt((1/m^2)*sum( (h(x) - mean(h(x)))^2 ))
}

integrals = NULL
variance_upper = NULL
variance_lower = NULL

ms = seq(from = 100, to = 20000, length = 300)
for (m in ms) {
  x = runif(m)
  integrals = c(integrals, mean(h(x)))
  variance_upper = c(variance_upper, mean(h(x)) + v(x, m))
  variance_lower = c(variance_lower, mean(h(x))-v(x, m))
}
```

```r
x = runif(1000)
mean(h(x))
mean(v(x, 1000))
```

```r
plot(ms, integrals, col = 'black')
lines(ms, variance_upper, col = 'blue')
lines(ms, variance_lower, col = 'red')
lines(ms, rep(0.965, m), col = 'green')
```
