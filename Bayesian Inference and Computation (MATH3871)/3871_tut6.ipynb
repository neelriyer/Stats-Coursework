```r
# (d)
n = 1e4

prior = function(x) {
  (1/2)*dbeta(x, 2, 4) + (1/2)*dbeta(x, 1, 1)
}

post = function(x) {
  0.6061*dbeta(x, 6, 10) + 0.3939*dbeta(x, 5, 7)
}

x = seq(0, 1, length = n)
plot(x, post(x), col = 'black', type = 'l')
lines(x, prior(x), col = 'blue')
legend(0.6,3,legend=c("Prior","Posterior"),col=c('blue', 'black'),lty=c(1,1))
```
