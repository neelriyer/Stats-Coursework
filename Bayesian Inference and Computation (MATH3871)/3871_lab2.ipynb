```r
inverse = function(x) {
  if (x>=1/2 && x < 1) {
    return ((-1/2)*log(-2*x+2))
  }
  if (0 < x && x < 1/2) {
    return ((1/2)*log(2*x))
  }
}

pdf = function(x) {
  pdf_output = NULL
  for (i in 1:length(x)) {

    value = x[i]

    if (value < 0) {
    pdf_output = c(pdf_output, exp(2*value))
    }
    if (value >= 0) {
      pdf_output = c(pdf_output, exp(-2*value))
    }
  }
  pdf_output
}

N = 1e5
values = runif(N)
output = NULL

for (i in 1:N) {
  output = c(output, inverse(values[i]))
}

hist(output, prob = T)
x = seq(from = -10, to = 10, length = N)
lines(x, pdf(x), col = 'blue')
```

```r
inverse = function(x) {
  if(x>=0 && x<1) {
    return (-lambda * log(1-x))
  }
  if (x<0) {
    return (0)
  }
}

N = 1e4
lambda = 367
values = runif(N)
output = NULL

for (i in 1:length(values)) {
  output = c(output, inverse(values[i]))
}

hist(output, prob = T)
x = seq(from = 0, to = max(output), length = N)
lines(x, dexp(x, rate = 1/lambda), col = 'blue')
```
